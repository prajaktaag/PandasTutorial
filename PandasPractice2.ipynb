# merge, concat, join, append

```python
#concat, axis = 0 by deafult
p1 = df3.iloc[0:3, 1:3]
p2 = df3.iloc[3:6, 1:3]
pd.concat([p1,p2])
```

```python
p1 = df3.iloc[0:3, 0:2]
p2 = df3.iloc[3:6, 2:4]
pd.concat([p1,p2], sort = False) 
#sort = False, to ignore the warning when 
#cols dont align and join = outer
```

```python
#join is outer
p1 = df3.iloc[0:3, 0:2]
p2 = df3.iloc[0:3, [0]]
pd.concat([p1,p2], join = 'inner') 
```

```python
p1 = df3.iloc[0:3, 0:2]
p2 = df3.iloc[0:3, 2:4]
pd.concat([p1,p2],axis = 1) 
```

```python
p1 = df3.iloc[0:3, 0:2]
p2 = df3.iloc[2:4, 0:2]
pd.concat([p1,p2],axis = 1) 
```

```python
p1 = df3.iloc[0:3, 0:2]
p2 = df3.iloc[0:3, 5:]
pd.concat([p1,p2],axis = 1)
```

```python
p1 = df3.A[0:3]
p2 = df3.B[0:3]
temp = pd.concat([p1,p2], keys = ["p1", "p2" ])
temp
```

```python
#append
#Append rows of other to the end of caller, 
#returning a new object. Columns in other that are not 
#in the caller are added as new columns.

p1 = df3.iloc[0:3,0:4]
p2 = df3.iloc[3:6,0:4]
p1.append(p2)
```

```python
p1 = df3.iloc[0:3,0:4]
p2 = df3.iloc[1:3,0:4]
p1.append(p2)
```

```python
p1 = df3.iloc[0:3,0:3]
p2 = df3.iloc[3:6,3:6]
p1.append(p2, sort = False) #sort for same reason like merge
```

```python
#merge
p1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
p2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                     'value': [5, 6, 7, 8]})
```

```python
pd.merge(p1,p2, on = p1.index)
```

```python
pd.merge(p1,p2, on = p1.lkey)
```

```python
p1.merge(p2, on = p1.index)
```
