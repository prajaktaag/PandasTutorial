# Group By
splits df in groups based on criteria, each group is a df  
apply functions independently to each group  
combine results into Dataframe object  
dict like structure key : df


```python
oo.groupby('Edition')
```

```python
type(oo.groupby('Edition'))
```

```python
list(oo.groupby('Edition'))
```

```python
#iterate through groups
for group, value in oo.groupby("Edition"):
    print(group)
    print(value)
```

```python
type(value)
```

```python
type(group)
```

#group by computation  
size() : count of each group  
count():    
first()/last():  
mean():  
max()/min():  
agg() : multiple stats in one calculation per group

```python
oo.groupby("Medal").size()
```

```python
oo.groupby("Edition").size()
```

```python
oo.groupby(["Edition", "NOC", "Medal"]).agg(["min", "max", "count"])
```

```python
oo.groupby(["Edition", "NOC", "Medal"]).size()
```

```python
#dict like structure
temp = oo.groupby(["Edition", "NOC", "Medal"]).agg({"Edition" : 
            ["min", "max", "count"]})
temp
```

```python
oo.loc[oo.Athlete == "LEWIS, Carl"].groupby('Athlete').agg({"Edition" : 
            ["min", "max", "count"]})
```

# Reshaping
stack(): inner cols to rows, reshape the dataset  
unstack(): reverse of statck

```python
#Athletes winning medals in 100 and 200m event of 2008 olymics
temp = oo[(oo.Edition == 2008) & ((oo.Event == "100m") 
                          | (oo.Event == "200m"))]
temp
```

```python
# group by country, gender, discipline,event
g = temp.groupby(["NOC","Gender", "Discipline","Event"]).agg('size')
```

```python
g
```

```python
df = g.unstack(['Discipline', 'Event'])
df
```

# stack
returns df or series  
never nan or missing data as by default dropna flag true  
inner levels of stack function are sorted    
taller df --> use stack function  
wide df --> use unsack function  

```python
df.stack()
```

```python
df.stack(['Discipline'])
```

# Unstack

new level of column labels  

```python
df.unstack()
```

```python
df.unstack('Gender')
```

# Excercise 5

```python
# No of gold medals won by US male and female throughout 
#the history of olympics 

temp = oo[(oo.NOC == "USA") & (oo.Medal == 'Gold')].groupby(['Edition', 'Gender'
                                ]).agg('size')
temp 
```

```python
temp.unstack('Gender', fill_value=0).plot();
```

```python
# 5 athletes who have won most gold 
# medals in the history of the olympics
#if tie, consider no of silver and then bronze medals

temp = oo.groupby(['Athlete', 'Medal']).size().unstack('Medal', fill_value = 0)
```

```python
temp
```

```python
temp.sort_values(by = ['Gold', 'Silver', 'Bronze'], ascending  = False)[['Gold', 'Silver', 'Bronze']].head(5
                                                                                            ).plot(kind = 'bar')
```

```python
# 3. present the summary of total medals won by
# participating countries in 2008 olympics

temp = oo[oo.Edition == 2008].groupby(['NOC', 'Medal']).agg('size')
temp
```
